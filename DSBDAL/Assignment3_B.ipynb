```python
def species_statistics(dataframe):
    for species in dataframe["Species"].unique():
        print(f"\nStatistical details for species: {species}\n")
        stats = (
            dataframe[dataframe["Species"] == species]
            .drop(columns=["Species"])
            .describe()
            .transpose()
        )
        
        display(stats)
```

```python
species_statistics(df)
```

```python
std_df = df.groupby("Species").std()

std_df.plot(kind="bar", figsize=(8,5))
plt.title("Standard Deviation of Features by Species")
plt.ylabel("Standard Deviation")
plt.xticks(rotation=0)
plt.show()
```
