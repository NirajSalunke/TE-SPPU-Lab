```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Fare', bins=30, kde=True)
plt.title('Distribution of Ticket Prices (Fare)', fontsize=14, fontweight='bold')
plt.xlabel('Fare', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

sns.countplot(data=df, x='Survived', ax=axes[0, 0], palette='viridis')
axes[0, 0].set_title('Survival Count')
axes[0, 0].set_xticklabels(['Did not survive', 'Survived'])

sns.countplot(data=df, x='Pclass', hue='Survived', ax=axes[0, 1], palette='coolwarm')
axes[0, 1].set_title('Survival by Passenger Class')
axes[0, 1].legend(title='Survived', labels=['No', 'Yes'])

sns.histplot(data=df, x='Age', bins=20, kde=True, ax=axes[1, 0], color='coral')
axes[1, 0].set_title('Age Distribution')

sns.boxplot(data=df, x='Pclass', y='Fare', ax=axes[1, 1], palette='Set3')
axes[1, 1].set_title('Fare Distribution by Class')

plt.tight_layout()
plt.show()
```

```python
# Violin Plot
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='Pclass', y='Fare', hue='Survived', split=True, palette='muted')
plt.title('Fare Distribution by Class and Survival Status', fontsize=14, fontweight='bold')
plt.xlabel('Passenger Class', fontsize=12)
plt.ylabel('Fare', fontsize=12)
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
# Swarm Plot
plt.figure(figsize=(10, 6))
sns.swarmplot(data=df[df['Fare'] < 100], x='Embarked', y='Fare', hue='Pclass', palette='deep')
plt.title('Fare Distribution by Embarkation Port (Fare < 100)', fontsize=14, fontweight='bold')
plt.xlabel('Embarkation Port', fontsize=12)
plt.ylabel('Fare', fontsize=12)
plt.legend(title='Class', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
# Select numerical columns for correlation
numeric_cols = df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr()
sns.heatmap(numeric_cols, annot=True, cmap='coolwarm', center=0, fmt='.2f', 
            linewidths=1, square=True)
plt.title('Correlation Matrix of Numerical Features', fontsize=14, fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df[df['Survived']==0]['Age'].dropna(), label='Did not survive', 
            fill=True, alpha=0.5, color='red')
sns.kdeplot(data=df[df['Survived']==1]['Age'].dropna(), label='Survived', 
            fill=True, alpha=0.5, color='green')
plt.title('Age Distribution by Survival Status (KDE : Kernel Density Estimation)', fontsize=14, fontweight='bold')
plt.xlabel('Age', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()
plt.show()
```
