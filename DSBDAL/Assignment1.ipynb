```python
# First Lets use Copy of DF for the same
df_first = df.copy()
```

```python
#   1. Convert Sruvived and Pclass to Categorial
df_first['Survived'] = df_first['Survived'].astype('category')
df_first['Pclass'] = df_first['Pclass'].astype('category')
```

```python
df_first.head()
df_first.dtypes
```

```python
# 2. Convert Sex and Embarked to Categorial
df_second = df.copy()
df_second['Sex'] = df_second['Sex'].astype('category')
df_second['Embarked'] = df_second['Embarked'].astype('category')
```

```python
df_second.head()
```

```python
df_second.dtypes
```

```python
# Handling Missing Values
#   1. Fill Up Missing Age with Mean Value
#   2. Fill Up Missing Embarked wth Mode Value


# 1. Fill Up Missing Age with Mean Value
# Lets Make a Copy First
df_third = df.copy()
# Get the Mean of Age
mean_age = df_third['Age'].mean()
```

```python
# Fill the Age in Place with mean_age
df_third['Age'].fillna(mean_age, inplace=True)
```

```python
df_third
```

```python
# here in original the 888th entry with age had NaN
df
```

```python
#   2. Fill Up Missing Embarked wth Mode Value
# Get Mode of Embarked
mode_value = df_third['Embarked'].mode()
```

```python
# Fill the Mode of Embarked inplace where the cells has NaN
df_third['Embarked'].fillna(mode_value, inplace=True)
```

```python
# Convert Categorial Variables into Quantitative Variables
#  1. Sex

# Lets Create another Column
df_third['Sex_Binary'] = df_third['Sex'].map({'male': 0, 'female': 1})
# Here we mapped Binary Number as per Gender
```

```python
df_third.head()
```

```python
# One Hot Encoding for Sex and Embarked
df_one_encoded = pd.get_dummies(df_third, columns=['Sex', 'Embarked'], drop_first=True)
```
